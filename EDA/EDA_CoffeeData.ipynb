```python
df = df.drop(['Target'], axis=1)
```

```python
from datetime import timedelta
```

```python
x = df['brew time (min)']
def seconder(x):
    mins, secs = map(float, x.split(':'))
    td = timedelta(minutes=mins, seconds=secs)
    return td.total_seconds()

df['seconds'] = df['brew time (min)'].apply(seconder)
```
